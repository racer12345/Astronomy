```python
import numpy as np
from astropy.io import fits

G = fits.open('frame-u-006122-1-0013.fits')
```

Info about data

```python
G.info()
```

Main Details of Galaxy(NCG 5866)

```python
data = G[0].data
G[0].header
```

```python
print("Min:",np.min(data))
print("Max:",np.max(data))
print("Std:",np.std(data))
print("Mean:",np.mean(data))
```

```python
import matplotlib.pyplot as plt

plt.style.use("Solarize_Light2")
plt.imshow(data,cmap='gray')
plt.colorbar()
```

Histogram

```python
H = plt.hist(data.flat, bins= 500)
```

Visualization Galaxy(NCG 5866) in Graph

```python
from astropy.visualization import make_lupton_rgb

Nu = fits.open('frame-u-006122-1-0013.fits')
Ng = fits.open('frame-g-006122-1-0013.fits')
Ni = fits.open('frame-i-006122-1-0013.fits')
```

```python
u = Nu[0].data
g = Ng[0].data
i = Ni[0].data
```

```python
rgb = make_lupton_rgb(u,g,i, stretch=1.2,Q=10)
plt.imshow(rgb,origin ='lower')
```
